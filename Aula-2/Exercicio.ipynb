```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Global_Cybersecurity_Threats_2015-2024.csv")


print("Shape:", df.shape)
display(df.head())
print("\nDtypes:")
print(df.dtypes)


print("\nNaN por coluna:")
print(df.isna().sum())

display(df.select_dtypes(include="number").describe())

cat_cols = df.select_dtypes(exclude="number").columns.tolist()
if len(cat_cols) > 0:
    for c in cat_cols:
        print(f"\nValue counts: {c}")
        display(df[c].value_counts().head(10))

plt.figure()
plt.hist(df["Attack Type"], bins=20)
plt.title("...")
# plt.spin(45)
plt.show()

plt.figure()
df.groupby("Attack Type")["Financial Loss (in Million $)"].mean().plot(kind="bar")
plt.title("...")
plt.show()

plt.figure()
plt.scatter(df["Attack Type"], df["Financial Loss (in Million $)"])
plt.title("...")
plt.show()
```

## Insight 1 (padrão)

Ao explorar o dataset Iris, foi possível observar um padrão claro entre as medidas das pétalas e as espécies das flores. As flores da espécie setosa apresentam pétalas significativamente menores do que as demais, enquanto versicolor e virginica possuem valores maiores. Isso sugere que as características das pétalas são importantes para diferenciar as espécies.

## Insight 2 (qualidade dos dados)

Durante a etapa de EDA foi verificado que o dataset não possui valores ausentes (NaN) e que as colunas apresentam tipos de dados adequados para análise. Além disso, as distribuições das variáveis parecem consistentes, indicando que os dados possuem boa qualidade para estudos estatísticos e modelagem.

## Insight 3 (hipótese/modelo futuro)

Com base nos padrões observados, é possível levantar a hipótese de que um modelo de classificação supervisionada consiga identificar corretamente a espécie de uma flor utilizando apenas suas medidas (comprimento e largura das sépalas e pétalas). Algoritmos como árvore de decisão, k-NN ou regressão logística provavelmente apresentariam bons resultados devido à separação visível entre as classes.
